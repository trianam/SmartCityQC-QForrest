```python
import pandas as pd
```

```python
pm = pd.read_csv('PM2.5-sensors.csv')
pm
```

```python
pm.ffill()
```

```python
pm = pm[(pm['Time']>= '2024-08-01 00:00:00') & (pm['Time']<= '2024-10-08 21:00:00')].ffill()
pm
```

```python
time = pm['Time'].iloc[::6].reset_index(drop=True)
time
```

```python
pm_no_time = pm.drop('Time', axis=1).reset_index(drop=True)
pm_no_time
```

```python
window_size = 6
l = [pm_no_time.iloc[i:i+window_size].mean(axis=0) for i in range(0, len(pm_no_time), window_size)]
l = pd.concat(l, axis=1).T
l
```

```python
len(l), len(time)
```

```python
pm25 = pd.concat([time, l], axis=1, ignore_index=True)
```

```python
pm25.columns = ['Time', 'PM2.5_Sensor1', 'PM2.5_Sensor2', 'PM2.5_Sensor3']
pm25
```

```python
pm = pd.read_csv('PM10-sensors.csv')
pm
```

```python
pm.ffill()
```

```python
pm = pm[(pm['Time']>= '2024-08-01 00:00:00') & (pm['Time']<= '2024-10-08 21:00:00')].ffill()
pm
```

```python
time = pm['Time'].iloc[::6].reset_index(drop=True)
time
```

```python
pm_no_time = pm.drop('Time', axis=1).reset_index(drop=True)
pm_no_time
```

```python
window_size = 6 
l = [pm_no_time.iloc[i:i+window_size].mean(axis=0) for i in range(0, len(pm_no_time), window_size)]
l = pd.concat(l, axis=1).T
l
```

```python
len(l), len(time)
```

```python
pm10 = pd.concat([time, l], axis=1, ignore_index=True)
```

```python
pm10.columns = ['Time', 'PM10_Sensor1', 'PM10_Sensor2', 'PM10_Sensor3']
pm10
```

```python
pm_total = pd.merge(pm25, pm10, on='Time')
pm_total
```

```python
pm_total.isna().any()
```

```python
pm_total.to_csv('PM_from_01-08-T-00-00_to_10-08-T-21-0@1hr.csv', index=False)
```
