```python
import pandas as pd
```

```python
df = pd.read_csv('unique_attendance_15/presenza_tot_finale.csv')
pm = pd.read_csv('particulate_matter/PM_final.csv')
weather = pd.read_csv('weather/weather-final-data.csv')
```

```python

df0 =df[df['areaAnalisi'].str.contains('Intero Comune')]
df0['areaAnalisi'] = 0
df1 =df[df['areaAnalisi'].str.contains('Cagliari - 001')]
df1['areaAnalisi'] = 1
df2 =df[df['areaAnalisi'].str.contains('Cagliari - 002')]
df2['areaAnalisi'] = 2
df3 =df[df['areaAnalisi'].str.contains('Cagliari - 003')]
df3['areaAnalisi'] = 3
df4 =df[df['areaAnalisi'].str.contains('Cagliari - 004')]
df4['areaAnalisi'] = 4
df5 =df[df['areaAnalisi'].str.contains('Cagliari - 005')]
df5['areaAnalisi'] = 5
df6 =df[df['areaAnalisi'].str.contains('Cagliari - 006')]
df6['areaAnalisi'] = 6
df7 =df[df['areaAnalisi'].str.contains('Cagliari - 007')]
df7['areaAnalisi'] = 7
df8 =df[df['areaAnalisi'].str.contains('Cagliari - 008')]
df8['areaAnalisi'] = 8
df9 =df[df['areaAnalisi'].str.contains('Cagliari - 009')]
df9['areaAnalisi'] = 9
df10 =df[df['areaAnalisi'].str.contains('Cagliari - 010')]
df10['areaAnalisi'] = 10
df11 =df[df['areaAnalisi'].str.contains('Cagliari - 011')]
df11['areaAnalisi'] = 11
df12 =df[df['areaAnalisi'].str.contains('Cagliari - 012')]
df12['areaAnalisi'] = 12

datetime = df0['datetime']
dfs = [df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12]
dfs = [d.drop(['datetime', 'areaAnalisi'], axis=1).reset_index(drop=True) for d in dfs]
attendance = pd.concat([datetime] + dfs, axis=1, ignore_index=True)
attendance
```

```python
attendance.columns = ['datetime'] + [f'AttendanceArea{i+1}' for i in range(12)] 
attendance
```

```python
final_ds = pd.concat([attendance, pm.drop('Time', axis=1), weather.drop('ts_get', axis=1)], axis=1)
final_ds
```
