```python
dfs = [df0, df1, df2, df3, df4, df5, df6, df7, df8, df9, df10, df11, df12]
```

```python
start = '2024-08-01 00:00:00'
stop = '2024-10-08 21:00:00'
```

```python
dfs = [df[(df['datetime'] >= start) & (df['datetime'] <= stop)].sort_values('datetime').reset_index(drop=True) for df in dfs]
dfs[0]
```

```python
def process(pm):
    columns = pm.columns
    window_size = 4
    time = pm['datetime'].iloc[::window_size].reset_index(drop=True)
    area = pm['areaAnalisi'].iloc[::window_size].reset_index(drop=True)
    pm.drop('areaAnalisi', inplace=True, axis=1)
    pm_no_time = pm.drop('datetime', axis=1).reset_index(drop=True)
    l = [pm_no_time.iloc[i:i+window_size].mean(axis=0) for i in range(0, len(pm_no_time), window_size)]
    l = pd.concat(l, axis=1).T
    pm10 = pd.concat([time, l, area], axis=1, ignore_index=True)
    pm10.columns = columns
    return pm10
```

```python
dfs = [process(df) for df in dfs]
```

```python
dfs[0]
```

```python
df = pd.concat(dfs, axis=0, ignore_index=True)
df
```

```python
df.to_csv(f'presenza_tot_from_{start}_to_{stop}@1hr.csv', index=False)
```
