```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Read the CSV file
# Replace 'your_file.csv' with your actual file path
dfw = pd.read_csv('weather-010824-061024.csv')
```

```python
dfw = dfw[dfw['ts_get'] >= '2024-08-01']
dfw = dfw[dfw['ts_get'] <= '2024-10-08 21:10:00']
dfw
```

```python
dfw['ts_get'] = dfw['ts_get'].apply(lambda x: x[:-3])
dfw
```

```python
ts = dfw['ts_get']
ts = sorted(list(set(ts)))
ts = [x.split(' ') for x in ts]
hashmap = {}
for x in ts:
    if x[0] not in hashmap:
        hashmap[x[0]] = 0
    hashmap[x[0]] += 1

for key, val in hashmap.items():
    if val != 24:
        print(key, val)
```
